```python
import os
from langchain.vectorstores import FAISS
from langchain.embeddings import SentenceTransformerEmbeddings
from langchain.document_loaders import UnstructuredFileLoader
from langchain.text_splitter import CharacterTextSplitter
from groq import Groq
from langchain.chains import RetrievalQA
```

```python
import requests
from bs4 import BeautifulSoup
from langchain.schema import Document

# List of QGIS URLs to load
qgis_urls = [
    "https://docs.qgis.org/latest/en/docs/user_manual/processing_algs/qgis/vectorgeometry.html#clip",
    "https://docs.qgis.org/latest/en/docs/user_manual/processing_algs/qgis/vectorgeometry.html#buffer",
    "https://docs.qgis.org/latest/en/docs/user_manual/processing_algs/qgis/vectorgeneral.html#reproject-layer",
    "https://docs.qgis.org/latest/en/docs/user_manual/processing_algs/qgis/vectoroverlay.html#intersect",
]

documents = []

for url in qgis_urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    text = soup.get_text(separator="\n", strip=True)

    documents.append(Document(page_content=text, metadata={"source": url}))
```

```python
# Split into chunks
text_splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=50)
docs = text_splitter.split_documents(documents)
```

```python
# Embed chunks
embedding_model = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")
vectorstore = FAISS.from_documents(docs, embedding_model)
```

```python
import os
from groq import Groq
from langchain.vectorstores import FAISS
from langchain.embeddings import SentenceTransformerEmbeddings
from langchain.text_splitter import CharacterTextSplitter
from langchain.schema import Document

with open("plan.txt", "r") as f:
    llm1_text = f.read()

data_schema = {
    "kerala_boundary": "data/kerala_boundary.shp",
    "dem_raw": "data/srtm_dem.tif",
    "dem_clipped": "outputs/kerala_dem.tif",
    "rivers_raw": "data/india_rivers.shp",
    "rivers_clipped": "outputs/kerala_rivers.shp",
    "waterbodies_raw": "data/india_waterbodies.shp",
    "waterbodies_clipped": "outputs/kerala_water.shp",
    "low_elevation_raster": "outputs/low_elevation.tif",
    "proximity_raster": "outputs/proximity_water.tif",
    "slope_raster": "outputs/slope_raster.tif",
    "flood_risk_raster": "outputs/flood_risk_raster.tif",
    "flood_risk_vector": "outputs/flood_risk_vector.shp",
    "final_map_project": "outputs/final_map.qgs"
}

# Setup Groq SDK
client = Groq(api_key="")

# Setup vector retriever
retriever = vectorstore.as_retriever()

# Step 3: Define query
query = "Identify flood-prone regions in Kerala"

# Step 4: Retrieve top documents
relevant_docs = retriever.get_relevant_documents(query)

# Step 5: Construct context
context = "\n\n".join(doc.page_content[:1000] for doc in relevant_docs)

prompt = f"""
You are a GIS workflow structuring agent.

Given a geospatial workflow plan (from Agent 1) and QGIS scripting context, return a detailed JSON array of execution steps.

Each item must contain:
- "step_id": Unique ID
- "name": Short label
- "description": One-line purpose of the step
- "algorithm": QGIS algorithm name (e.g., "gdal:cliprasterbymasklayer")
- "input_files": List of input file paths (from the data schema below)
- "output_file": Output file path (also from the data schema)
- "script_filename": Python filename that will be generated for this step

--- BEGIN DATA_SCHEMA ---
{json.dumps(data_schema, indent=2)}
--- END DATA_SCHEMA ---

--- BEGIN QGIS CONTEXT ---
{context}
--- END QGIS CONTEXT ---

--- BEGIN WORKFLOW PLAN (FROM AGENT 1) ---
{llm1_text}
--- END WORKFLOW PLAN ---

Do NOT include any extra explanation. Output must be a clean JSON array of step objects only.
"""

response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {"role": "system", "content": "You are a precise geospatial Python workflow planner."},
        {"role": "user", "content": prompt}
    ]
)

code_plan = response.choices[0].message.content.strip()

with open("code_plan.json", "w") as f:
    f.write(code_plan)

print("✅ Saved structured plan to code_plan.json")

```
